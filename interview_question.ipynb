```python
# check if input number is a leap year or not . If leap , check if the number is palindrome or not.
# You cannot convert number to string.
# 1600->l

# 2000-> l

# 2012->l

# 1900-> nl

# 1800->nl

# 1804->l

# 2112-> l and palindrome
```

```python
n = 2112
temp = n
res = 0
while (n!=0):
  res = res*10+(n%10)
  n = n//10

if temp == res:
  print('palindrome')
else:
  print('not palindrome')
```

```python
# [1, 2, 3, 1] ⇒ [1, 2, 3]
# [1, 3, 2, 1, 5, 3, 5, 1, 4] ⇒ [1, 3, 2, 5, 4]
```

```python
def removeDuplicate(lst):
  res = []
  for i in lst:
    if i not in res:
      res.append(i)
  return res
```

```python
lst = [1, 3, 2, 1, 5, 3, 5, 1, 4]

print(removeDuplicate(lst))
```

```python
# prime no
# n = 11

def isPrime(n):
  for i in range(1,n):
    if i>1:
      for n in range(2,i):
        if (i%n)==0:
          break
        else:
          print(i)
n = 11
print(isPrime(n))
```
